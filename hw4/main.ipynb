```python
from math import sqrt
import random
import matplotlib.pyplot as plt
import numpy as np

def PDF_function(x):
    return(2*x)

def CDF_function(x):
    return(x*x)

x = np.linspace(0, 1, 20) 
y_PDF = []
y_CDF = []

for i in x:
    y_PDF.append(PDF_function(i))
    y_CDF.append(CDF_function(i))
    
def invers_CDF_function(x):
    return sqrt(x)

random_numbers = []
distribution = []

for _ in range(1000):
    random_numbers.append(random.random())
    
for i in random_numbers:
    distribution.append(invers_CDF_function(i))

fig, ax = plt.subplots() 
ax.plot(x, y_PDF, color="green", label="PDF")
ax.plot(x, y_CDF, color="red", label="CDF")
ax.hist(distribution, bins = 30, density=True, cumulative = True, alpha=0.5, label="гистограмма распределения")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.rcParams["figure.figsize"] = (8,5)
plt.savefig('increasing.png')
plt.show() 
```

```python
from math import sqrt
import random
import matplotlib.pyplot as plt
import numpy as np

    
def PDF_function(x):
    if x > 0.5:  
        return(2-2*x)
    else:
        return 2*x

def CDF_function(x):
    if x > 0.5:
        return (-x*(x-2))
    else:
        return x*x

x = np.linspace(0, 1, 100) 
y_PDF = []
y_CDF = []

for i in x:
    y_PDF.append(PDF_function(i))
    y_CDF.append(CDF_function(i))
    
def invers_CDF_function(x):
    if x > 0.5:
        return (1-sqrt(1-x))
    else:
        return sqrt(x)

random_numbers = []
distribution = []

for _ in range(1000):
    random_numbers.append(random.random())
    
for i in random_numbers:
    distribution.append(invers_CDF_function(i))


fig, ax = plt.subplots() 
ax.plot(x, y_PDF, color="green", label="PDF")
ax.plot(x, y_CDF, color="red", label="CDF")
ax.hist(distribution, bins = 30, density=True, cumulative = True, alpha=0.5, label="гистограмма распределения")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.rcParams["figure.figsize"] = (8,5)
ax.legend()
plt.savefig('triangular.png')
plt.show() 
```
